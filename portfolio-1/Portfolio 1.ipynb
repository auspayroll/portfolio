# Analysis of Cycling Data

## Loading Data

This is an analysis of a cyclists ride data exported from [Strava](https://strava.com/), an online social 
network site for cycling and other sports.  The dataset is a log of every ride since the start of 2018
and contains summary data like the distance and average speed.  It was exported using 
the script `scripts/stravaget.py` which uses the stravalib module to read data. Some details of
the fields exported by that script can be seen in [the documentation for stravalib](https://pythonhosted.org/stravalib/api.html#stravalib.model.Activity). 


```python
strava = pd.read_csv('data/strava_export.csv', index_col='date', parse_dates=True)
strava.head(50)
```

The second dataset comes from an application called [GoldenCheetah](https://www.goldencheetah.org/) which provides
some analytics services over ride data.  This has some of the same fields but adds a lot of analysis of the 
power, speed and heart rate data in each ride.  This data overlaps with the Strava data but doesn't include all 
of the same rides. 

We create an index using the datetime for each ride, this time combining two columns in the data (date and time) 
and localising to Sydney so that the times match those for the Strava data. 

```python
cheetah = pd.read_csv('data/cheetah.csv', skipinitialspace=True)
cheetah.index = pd.to_datetime(cheetah['date'] + ' ' + cheetah['time'])
cheetah.index = cheetah.index.tz_localize('Australia/Sydney')
cheetah.head()
```

Here are definitions of some of the more important fields in the data. Capitalised fields come from the GoldenCheetah data
while lowercase_fields come from Strava. There are many cases where fields are duplicated and in this case the values
should be the same, although there is room for variation as the algorithm used to calculate them could be different
in each case. 

  * Duration - overall duration of the ride, should be same as elapsed_time
  * Time Moving - time spent moving (not resting or waiting at lights), should be the same as moving_time
  * Elevation Gain - metres climbed over the ride
  * Average Speed - over the ride
  * Average Power - average power in watts as measured by a power meter, relates to how much effort is being put in to the ride, should be the same as  * average_watts' from Strava
  * Nonzero Average Power - same as Average Power but excludes times when power is zero from the average
  * Average Heart Rate - should be the same as average_heartrate
  * Average Cadence - cadence is the rotations per minute of the pedals
  * Average Temp - temperature in the environment as measured by the bike computer (should be same as average_temp)
  * VAM - average ascent speed - speed up hills
  * Calories (HR) - Calorie expendature as estimated from heart rate data
  * 1 sec Peak Power - this and other  'Peak Power' measures give the maximum power output in the ride over this time period.  Will be higher for shorter periods. High values in short periods would come from a very 'punchy' ride with sprints for example.
  * 1 min Peak Hr - a similar measure relating to Heart Rate
  * NP - Normalised Power, a smoothed average power measurement, generally higher than Average Power 
  * TSS - Training Stress Score, a measure of how hard a ride this was
  * device_watts - True if the power (watts) measures were from a power meter, False if they were estimated
  * distance - distance travelled in Km
  * kudos - likes from other Strava users (social network)
  * workout_type - one of  'Race',  'Workout' or  'Ride'
  
  
Some of the GoldenCheetah parameters are defined [in thier documentation](https://github.com/GoldenCheetah/GoldenCheetah/wiki/UG_Glossary).  

### Merging datasets

The datasets from Strava and GoldenCheetah are merged by an inner join on the timestamp indexes from each. 

```python
#left_on: use strava.date as key; left_index=True, right_idex=True: use indexes to join; 
#default is inner join
ds = pd.merge(strava, cheetah, left_index=True, right_index=True)
ds
```

Rides with no measured power are removed (device_watts is False) - these are commutes or MTB rides. 

```python
dsF = ds[ds.device_watts != False]
```

## 2. Distributions of covariates

Below, we will inspect the distributions of some key variables: time, distance, average speed, average power, Training Stress Score (TSS). 

### Time 

For the time distribution, 'time in motion' or 'moving time is used'. The distribution of moving time seems to be around 2 ( or even 3) main clusters. Two right skewed clusters are centered around 60 minutes and approximately 140 minutes. Another smaller clusters appears around 15-20 minutes. 

```python
time = dsF['moving_time'].hist(bins=20)
time.set_xlabel("Moving Time (minutes)")
```

### Distance

The distribution of distance appears very similar to the distribution of time in motion. This makes sense as time moving translates directly to distance travelled. The correlation between time moving and distance is 0.93. There are high frequencies around 10, 30, 60 distance units. There is also a smaller outlier in excess of 100 units.  

```python
distance = dsF['Distance'].hist(bins=20)
distance.set_xlabel("Distance")
```

### Speed 
The speed distribtion appears to have 3 main clusters, a smaller one around 5km per hour, a larger normally distributed cluster around 25km per hour, and a moderate cluster approximately normally distributed around 35/36 km per hour. 

```python
speed = dsF['Average Speed'].hist(bins=20)
speed.set_xlabel("Speed")
```

### Power
Power in watts appears to be normally distributed around 150 watts (right skewed). There are two main peaks in this cluster, one around 150 watts, and another around 200 watts. This could suggest that there are two clusters close together. An additional smaller cluster exists with very low power ~ 5 - 10 watts. 

```python
power = dsF['Average Power'].hist(bins=20)
power.set_xlabel("Power")
```

### Training Stress Score
Again, there appears to be 3 distinctive clusters, 2 normally distributed right skewed around 20 and 180 respectively, and additional smaller cluster around 400. 

```python
tss = dsF['TSS'].hist(bins=20)
tss.set_xlabel("Training stress score")
```

### Summary

Among the distributions for time, distance, speed, power and training stress score, there appear to be around 2, mostly 3 distinctive clusters. This would seem to indicate three levels or intensity of rides. 

## 3. Relationships between covariates

Using the correlation matrix below, the relationships between key covariates will be explored.

As expected, there is a high correlation 'Time Moving' and 'Distance', aswell as between 'Distance' and 'Elevation gain' (0.933 and 0.805 respectively). 

With respect to training stress score, 3 factors are highly correlated: elevation gain, distance and time moving. Since distance and time moving are highly dependant/interrelated, only one of these would be considered as covariates for a proposed model. Surprisingly, heart rate and average power are not highly correlated to training stress score (0.114 and 0.23 respectively).

Average heart rate and average power were moderately correlated at 0.69, as was heart rate and average speed at 0.74.

As expected, average power and normalised power were highly correlated (0.844).

```python
dsF1 = dsF.loc[:,['Distance', 'Time Moving', 'Average Speed', 'average_heartrate', 'NP', 'Elevation Gain', 'TSS', 'Average Power']]
dsF1
cmatrix = dsF1.corr()
cmatrix
```

## 5. Monthly Summary

```python
summary = dsF.resample('m').agg({'distance': 'sum', 'TSS': 'sum', 'Average Speed': 'mean'})
summary = summary.assign(month=summary.index.strftime("%b %Y")).set_index('month')

# make a plot with different y-axis using second axis object
fig,ax = plt.subplots()
ax2=ax.twinx()

summary.distance.plot(kind='bar', ax=ax, color="C1", position=0, label="Total Distance")
summary.TSS.plot(kind='bar', ax=ax, position=1, color="C6", label="Total TSS")
summary['Average Speed'].plot(ax=ax2, color="C2", marker="o")

fig.set_size_inches(15, 6)
ax2.set_ylabel("Speed km/h")
ax2.grid(False)
fig.legend()
```

### Daily activity summary

Below is a daily activity summary for July 2019, showing total distance, total training test score and average speed for each day. 

```python
import datetime
summary = dsF[(dsF.index.month == 3) & (dsF.index.year == 2019)].resample('d').agg({'distance': 'sum', 'TSS': 'sum', 'Average Speed': 'mean'})
summary.loc[summary['Average Speed'].isna(), ['Average Speed']] = 0
summary = summary.assign(day=summary.index.strftime("%-d")).set_index('day')


# make a plot with different y-axis using second axis object
fig,ax = plt.subplots()
ax.set_ylabel("Distance, TSS")
ax2=ax.twinx()

summary.distance.plot(kind='bar', ax=ax, color="C1", position=0, label="Total Distance", width=0.4)
summary.TSS.plot(kind='bar', ax=ax, position=1, color="C6", label="Total TSS", width=0.4)
summary['Average Speed'].plot(kind='bar', ax=ax2, color="C2", width=0.3)

fig.set_size_inches(15, 6)
ax2.set_ylabel("Speed km/h")
ax2.grid(False)
fig.legend()
```
